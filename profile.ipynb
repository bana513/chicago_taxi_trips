```python
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
df = pd.read_csv("taxi_1m.csv")
```

```python
df.head()
```

```python
df_profile = df.sample(n=10000, replace=False).reset_index(drop=True).copy()
```

```python
# Create profiling
profile = ProfileReport(df_profile)

# Save report to file
profile.to_file("report.html")
```
